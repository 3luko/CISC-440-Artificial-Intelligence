### Toggle Rule: Neighborhood Toggle

A move chooses one cell (by index or (row, col)) and flips:

- the chosen cell itself
- its **up** neighbor (if it exists)
- its **down** neighbor (if it exists)
- its **left** neighbor (if it exists)
- its **right** neighbor (if it exists)

Flipping means:

- `0 ‚Üí 1`
- `1 ‚Üí 0`

**No diagonals**.  
If a neighbor is outside the grid (edge/corner), ignore it.

### Example: 2√ó2 Indexing

Use row-major indexing:

0 1
2 3

- Toggle `0` affects cells `{0, 1, 2}`
- Toggle `1` affects cells `{1, 0, 3}`
- Toggle `2` affects cells `{2, 0, 3}`
- Toggle `3` affects cells `{3, 1, 2}`
