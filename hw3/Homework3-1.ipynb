# Problem 2 — Tower of Hanoi

## Scenario

We want to move a stack of **N disks** from a start peg to a goal peg using an auxiliary peg.

### Rules

- Only one disk may be moved at a time.
- Only the **top disk** of any peg may be moved.
- A larger disk may **never** be placed on top of a smaller disk.

## Example (N = 3)

Initial state:

    A = (3, 2, 1)
    B = ()
    C = ()

Goal state:

    A = ()
    B = ()
    C = (3, 2, 1)

## Your Task

Formulate this as a search problem by defining the following components.

### State Space

- What is a state?
- How do you represent the peg configuration?

### Initial State

For N disks starting on peg 0:

### Goal State

All disks stacked on the goal peg (e.g., peg 2):

### Actions

- What actions are available from any state?

A move is legal only if:

1. The source peg is not empty.
2. The destination peg is empty OR its top disk is larger than the disk being moved.

### Transition Model

Applying action (i, j):

1. Remove the top disk from peg i.
2. Place it on peg j.
3. Return the updated state.

### Path Cost

Each move costs 1.

Mathematically:

    g(child) = g(parent) + 1

The total cost equals the number of moves.


```python
from search import Problem

class Hanoi(Problem):

    def __init__(self, N):
        self.N = N
        initial = (tuple(range(N, 0, -1)), (), ())
        goal = ((), (), tuple(range(N, 0, -1)))
        super().__init__(initial, goal)
        
    # Actions
    def actions(self, state):
        moves = []
        for i in range(3):
            for j in range(3):
                if i != j and self.is_legal(state, i, j):
                    moves.append((i, j))
        return moves

    def is_legal(self, state, i, j):
        if not state[i]:
            return False
        if not state[j]:
            return True
        return state[i][-1] < state[j][-1]
   
    # Transition model
    def result(self, state, action):
        i, j = action
        new_state = [list(peg) for peg in state]

        disk = new_state[i].pop()
        new_state[j].append(disk)

        return tuple(tuple(peg) for peg in new_state)
    
    # goal test
    def goal_test(self, state):
        return state == self.goal
    
    # Path cost
    def path_cost(self, c, state1, action, state2):
        return c + 1
```

# Problem 3 — Toggle-It Puzzle

### Scenario
You are given an **N×N grid** of binary values:

- `0` = OFF
- `1` = ON

Your goal is to transform the grid from an initial state to a goal state
(typically **all zeros**).

### Toggle Rule: Neighborhood Toggle

A move chooses one cell (by index or (row, col)) and flips:

- the chosen cell itself
- its **up** neighbor (if it exists)
- its **down** neighbor (if it exists)
- its **left** neighbor (if it exists)
- its **right** neighbor (if it exists)

Flipping means:

- `0 → 1`
- `1 → 0`

**No diagonals**.  
If a neighbor is outside the grid (edge/corner), ignore it.

### Example: 2×2 Indexing

Use row-major indexing:

0 1
2 3

- Toggle `0` affects cells `{0, 1, 2}`
- Toggle `1` affects cells `{1, 0, 3}`
- Toggle `2` affects cells `{2, 0, 3}`
- Toggle `3` affects cells `{3, 1, 2}`


### State Representation
Represent a state as a binary string of length `N*N`:

For a 2×2 grid:

- `"0011"` represents:

00

11


- Goal `"0000"` represents:


00

00


### Search Problem Components

#### State Space
All binary grids of size N×N (equivalently all strings of length N² over {0,1}).

#### Initial State
Example: `"0011"`

#### Goal State
Example: `"0000"`


#### Actions

At any state, an action consists of choosing one cell index:

action ∈ {0, 1, …, N² − 1}

Therefore, each state has exactly N² possible actions.


#### Transition Model
Applying action `i` flips the chosen cell and its valid neighbors (up/down/left/right),
producing a new grid configuration.

#### Path Cost
Each toggle costs 1


## Expected Output Style (Example)

**Testing 2x2 problem**  
Solving `0011 => 0000` 

Example of a valid solution trace:

initial

00

11

Toggle 0 =>

11

01

Toggle 1 =>

00

00

```python
from search import Problem
class ToggleItProblem(Problem):

    def __init__(self, size, initial, goal=None):
        self.size = size
        self.N = size * size

        if goal is None:
            goal = "0" * self.N

        super().__init__(initial, goal)

    # Actions
    def actions(self, state):
        # Every cell index is always a valid action
        return list(range(self.N))

    # Transition model
    def result(self, state, action):
        state_list = list(state)
        r, c = divmod(action, self.size)

        # Cells to flip
        cells = [(r, c)]

        if r > 0:                     # up
            cells.append((r - 1, c))
        if r < self.size - 1:         # down
            cells.append((r + 1, c))
        if c > 0:                     # left
            cells.append((r, c - 1))
        if c < self.size - 1:         # right
            cells.append((r, c + 1))

        # Flip bits
        for row, col in cells:
            idx = row * self.size + col
            state_list[idx] = '1' if state_list[idx] == '0' else '0'

        return "".join(state_list)

    # Goal test
    def goal_test(self, state):
        return state == self.goal

    # Path cost
    def path_cost(self, c, state1, action, state2):
        return c + 1

    
    # Display
    def display(self, state, action=None):
        if action is not None:
            print(f"(Toggled index {action})")

        for i in range(self.size):
            row = state[i * self.size:(i + 1) * self.size]
            print(" ".join(row))
        print()
```
